```python
from classy import Class
```

```python
cosmo=Class()
```

```python
cosmo.set({'N_eff':3.046,'output':'tCl'})
#cosmo.Hubble(0)
```

```python
cosmo.compute()
```

```python
cl=cosmo.raw_cl(2500)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
l = np.array( range (2,2501))
factor = l*(l+1)/(2*np.pi)
cosmo.set({'N_eff':3.046})
cosmo.compute()
cl2=cosmo.raw_cl(2500)
plt.plot(np.arange(2,2501),cl['tt'][2:]/cl2['tt'][2:])
```

```python

cosmo2=Class()
cosmo2.Hubble(0)
```

```python
cosmo.set({'N_eff':3.146})
cosmo.compute()
cl3=cosmo.raw_cl(2500)
```

```python
plt.plot(np.arange(2,2501),cl['tt'][2:]/cl3['tt'][2:])
```

```python
plt.plot(np.arange(2,2501),cl2['tt'][2:]/cl['tt'][2:])
```

```python
cosmo.set({'N_eff':4})
cosmo.compute()
cl=cosmo.raw_cl(2500)
cosmo.set({'N_eff':2})
cosmo.compute()
cl3=cosmo.raw_cl(2500)
plt.loglog(l,factor*cl['tt'][2:],label='larger')
plt.loglog(l,factor*cl2['tt'][2:],label='default')
plt.loglog(l,factor*cl3['tt'][2:],label='smaller')
plt.xlim([100,2500])
plt.legend()
```

```python
cosmo.set({'h':0.8})
cosmo.compute()
cl=cosmo.raw_cl(2500)
cosmo.set({'h':0.65})
cosmo.compute()
cl3=cosmo.raw_cl(2500)
cosmo.set({'h':0.7})
cosmo.compute()
cl2=cosmo.raw_cl(2500)
plt.loglog(l,factor*cl['tt'][2:],label='larger')
plt.loglog(l,factor*cl2['tt'][2:],label='default')
plt.loglog(l,factor*cl3['tt'][2:],label='smaller')
plt.xlim([100,2500])
plt.legend()
```
