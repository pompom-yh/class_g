```python
cosmo1c_ad=Class()
```

```python
cosmo1c_ad.set({'output':'tCl','l_max_scalars':4000})
cosmo1c_ad.compute()
```

```python
tt_ad=cosmo1c_ad.raw_cl()['tt']
```

```python
cosmo1c_cdi=Class()
cosmo1c_cdi.set({'output':'tCl','ic':'ad&cdi','f_cdi':2,'n_cdi':1,'l_max_scalars':4000})
cosmo1c_cdi.compute()
```

```python
tt_cdi=cosmo1c_cdi.raw_cl()['tt']
```

```python
cosmo1c_nid=Class()
cosmo1c_nid.set({'output':'tCl','ic':'ad&nid','f_nid':4,'n_nid':1,'l_max_scalars':4000})
cosmo1c_nid.compute()
```

```python
tt_nid=cosmo1c_nid.raw_cl()['tt']
```

```python
l=np.arange(4001)
ll=l*(l+1)/2/np.pi
```

```python
plt.plot(l,ll*tt_ad,label='ad')   # in class python, different ic mode outputs are summed up, while in c, are separate files.
plt.plot(l,ll*(tt_cdi-tt_ad),label='cdi')
plt.plot(l,ll*(tt_nid-tt_ad),label='nid')
plt.xscale('log')
plt.xlim((2,4000))
plt.ylim((0,1e-9))
plt.legend()
```
